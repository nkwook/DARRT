```python
import os
import shutil


SOURCE_DIR_INSIDE_DATA = "./data/SPARK-data"
DEST_DIR = "./data"

if os.path.exists(SOURCE_DIR_INSIDE_DATA) and os.path.isdir(SOURCE_DIR_INSIDE_DATA):
    for item_name in os.listdir(SOURCE_DIR_INSIDE_DATA):
        s_item = os.path.join(SOURCE_DIR_INSIDE_DATA, item_name)
        d_item = os.path.join(DEST_DIR, item_name)

        if os.path.exists(d_item):
            print(f"'{d_item}'is already exsits")

        else:
            shutil.move(s_item, DEST_DIR)
            print(f"'{s_item}' -> '{d_item}' done.")

    if not os.listdir(SOURCE_DIR_INSIDE_DATA):
        print(f"'{SOURCE_DIR_INSIDE_DATA}' folder is empty")
        os.rmdir(SOURCE_DIR_INSIDE_DATA)
    else:
        print(f"'{SOURCE_DIR_INSIDE_DATA}' folder has a file yet")
else:
    print(f"ERROR : We can not find folder '{SOURCE_DIR_INSIDE_DATA}'")
```

```python

DATASET_NAME = "icews14"
LLM_MODEL_HF_ID = "meta-llama/Llama-2-7b-hf"
RAG_TYPE = "TLR" # or ICL
ADAPTER_NAME = "xERTE" # "Tlogic"

# -------------------------------------------------------------

LLM_MODEL_NAME_FOR_PATH = LLM_MODEL_HF_ID.split('/')[-1]
DRIVE_LLM_CACHE_DIR = f"{DRIVE_PROJECT_ROOT}/llm_cache/{LLM_MODEL_NAME_FOR_PATH}/{RAG_TYPE}"
LOCAL_LLM_RESULT_DIR_FOR_TRAINING = f"./data/llm_result/{LLM_MODEL_NAME_FOR_PATH}/{RAG_TYPE}"

!mkdir -p {LOCAL_LLM_RESULT_DIR_FOR_TRAINING}
if os.path.exists(DRIVE_LLM_CACHE_DIR) and any(os.scandir(DRIVE_LLM_CACHE_DIR)):
    !cp -r "{DRIVE_LLM_CACHE_DIR}/." "{LOCAL_LLM_RESULT_DIR_FOR_TRAINING}/"
    !ls -l "{LOCAL_LLM_RESULT_DIR_FOR_TRAINING}"
else:
    print(f"No LLM cahce in : {DRIVE_LLM_CACHE_DIR} !!!")

DRIVE_EMBEDDING_PATH = f"{DRIVE_PROJECT_ROOT}/data_for_vm/original/{DATASET_NAME}/ent_rel_emb.pt"
LOCAL_ORIGINAL_DATA_FOR_EMB_PATH = f"./data/original/{DATASET_NAME}"
!mkdir -p {LOCAL_ORIGINAL_DATA_FOR_EMB_PATH}
if os.path.exists(DRIVE_EMBEDDING_PATH):
    if not os.path.exists(f"{LOCAL_ORIGINAL_DATA_FOR_EMB_PATH}/ent_rel_emb.pt"):
        !cp "{DRIVE_EMBEDDING_PATH}" "{LOCAL_ORIGINAL_DATA_FOR_EMB_PATH}/"
    else:
        print(f"'{LOCAL_ORIGINAL_DATA_FOR_EMB_PATH}/ent_rel_emb.pt' is already in your VM.")
else:
    print(f"You should have ent_rel_emb.pt file!")  #if you have this file in your VM, is is okay. Then you can ignore this.

if not (os.path.exists(f"./data/original/{DATASET_NAME}/entity2id.json") and \
        os.path.exists(f"./data/processed/{DATASET_NAME}/{DATASET_NAME}.json")):
    print(f"You should set the dataset.")
else:
    print(f"(./data/original, ./data/processed) files are ready.")

# Trained model is saved in --OUTPUT_DIR
TRAINED_ADAPTER_DRIVE_DIR = f"{DRIVE_PROJECT_ROOT}/trained_adapters/{DATASET_NAME}_{LLM_MODEL_NAME_FOR_PATH}_{RAG_TYPE}_{ADAPTER_NAME}"
!mkdir -p "{TRAINED_ADAPTER_DRIVE_DIR}"


# if Tlogic you should modify. e.g.--RULE_DATA_PATH
!python main.py \
  --DATASET {DATASET_NAME} \
  --MODEL_NAME "{LLM_MODEL_HF_ID}" \
  --RAG "{RAG_TYPE}" \
  --GEN_MODE "beam" \
  --LOAD_LLM \
  --ADAPTER_NAME "{ADAPTER_NAME}" \
  --DEVICE "cuda:0" \
  --TRAIN_BS 64 \
  --EVAL_BS 64 \
  --EPOCHS 10 \
  --LEARNING_RATE 1e-4 \
  --OUTPUT_DIR "{TRAINED_ADAPTER_DRIVE_DIR}" \
  --WANDB False \
  --NUM_WORKERS 0 \
  --ORI_DATA_PATH "./data/original" \
  --DATA_PATH "./data/processed" \
  --RULE_DATA_PATH "./data/rule_output"

print(f"\n--- TKG adapter train is done ({ADAPTER_NAME}) ---")
print(f"Trained model is saved on : {TRAINED_ADAPTER_DRIVE_DIR}")
```

Set the environment for LLM-DA adapter. We only perform with icews14 dataset. If you want to perform with other dataset, you may modify the codes.

```python
import os

DATASET_NAME = "icews14"  # "icews18", "GDELT"

# To Duplicate the dataset from your local to VM, set the path
LLMDA_DATA_BASE_DIR = "./data/LLM-DA/"
LLMDA_VOCAB_DIR_IN_NOTEBOOK = os.path.join(LLMDA_DATA_BASE_DIR, DATASET_NAME)
LLMDA_RULES_INPUT_DIR_IN_NOTEBOOK = os.path.join(LLMDA_DATA_BASE_DIR, "ranked_rules", DATASET_NAME)

DRIVE_LLMDA_VOCAB_SOURCE_DIR = f"{DRIVE_PROJECT_ROOT}/LLM-DA/datasets/{DATASET_NAME}" # modify for your environment
DRIVE_LLMDA_RULES_SOURCE_DIR = f"{DRIVE_PROJECT_ROOT}/LLM-DA/ranked_rules/{DATASET_NAME}" # modify for your environment, in this folder you should have confidence.json file.
LLMDA_RULE_FILENAME = "confidence.json" # You should get this file by run the LLM-DA code, providing with this notebook file.

llmda_ent2id_drive_path = os.path.join(DRIVE_LLMDA_VOCAB_SOURCE_DIR, "entity2id.json")
llmda_rel2id_drive_path = os.path.join(DRIVE_LLMDA_VOCAB_SOURCE_DIR, "relation2id.json")

#We should convert the rulebanks for suitable at SPARK, since ent2id mappings are difference between SPARK and LLM-DA in same dataset icews14
CONVERTED_RULES_OUTPUT_DIR = f"./data/converted_rules_for_spark/{DATASET_NAME}/"


# Make folders. You should prepare the files(ent2id.json, rel2id.json) in appropriate folders.
if not os.path.exists(LLMDA_VOCAB_DIR_IN_NOTEBOOK):
    os.makedirs(LLMDA_VOCAB_DIR_IN_NOTEBOOK, exist_ok=True)

if not os.path.exists(LLMDA_RULES_INPUT_DIR_IN_NOTEBOOK):
    os.makedirs(LLMDA_RULES_INPUT_DIR_IN_NOTEBOOK, exist_ok=True)

# Converted rules (LLM-DA -> SPARK)
if not os.path.exists(CONVERTED_RULES_OUTPUT_DIR):
    os.makedirs(CONVERTED_RULES_OUTPUT_DIR, exist_ok=True)

if os.path.exists(llmda_ent2id_drive_path):
    !cp "{llmda_ent2id_drive_path}" "{LLMDA_VOCAB_DIR_IN_NOTEBOOK}/"
else:
    print(f"No LLM-DA ent2id file in : {llmda_ent2id_drive_path}")

if os.path.exists(llmda_rel2id_drive_path):
    !cp "{llmda_rel2id_drive_path}" "{LLMDA_VOCAB_DIR_IN_NOTEBOOK}/"
else:
    print(f"No LLM-DA rel2id file in : {llmda_rel2id_drive_path}")

# if you need ts2id file also, you can.
llmda_ts2id_drive_path = os.path.join(DRIVE_LLMDA_VOCAB_SOURCE_DIR, "ts2id.json")
if os.path.exists(llmda_ts2id_drive_path):
    !cp "{llmda_ts2id_drive_path}" "{LLMDA_VOCAB_DIR_IN_NOTEBOOK}/"
else:
    print(f" No ts2id file in : {llmda_ts2id_drive_path}")


llmda_rules_drive_path = os.path.join(DRIVE_LLMDA_RULES_SOURCE_DIR, LLMDA_RULE_FILENAME)

if os.path.exists(llmda_rules_drive_path):
    !cp "{llmda_rules_drive_path}" "{LLMDA_RULES_INPUT_DIR_IN_NOTEBOOK}/"
else:
    print(f"No rules(confidence.json) in : {llmda_rules_drive_path}")
    print(f"You should get LLM-DA rules first!")
```
